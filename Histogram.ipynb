```
import matplotlib as plt
import numpy as np
import random
ml_students_age = np.random.randint(18, 45, (100))
py_students_age = np.random.randint(15, 40, (100))
print(ml_students_age)
print(py_students_age)
```

```
import matplotlib.pyplot as plt

ml_students_age = np.random.randint(18, 45, (100))
py_students_age = np.random.randint(15, 40, (100))
print(ml_students_age)
print(py_students_age)

plt.hist(ml_students_age)

plt.title("ML students age")
plt.xlabel("Student age category")
plt.ylabel("No. of students")
plt.show()
```

```
from matplotlib import style
style.use('ggplot')
bins = [15, 20, 25, 30, 35, 40, 45]
plt.figure(figsize= (16, 9))

plt.hist([ml_students_age, py_students_age], bins, rwidth=0.8, histtype="bar", orientation="vertical", color= ['m', 'y'], label = ["ML students", "Py students"] )

plt.title("ML & Py students age histogram")
plt.xlabel("Student age category")
plt.ylabel("No. of students")
plt.show()
```
